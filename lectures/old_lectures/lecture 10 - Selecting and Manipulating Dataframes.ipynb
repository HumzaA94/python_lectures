```python
list_ = []
for c in df.columns:
    val = c.replace(" ","_")
    list_.append(val)
    
df.columns = list_
```

```python
df.columns = [c.replace(" ", "_") for c in df.columns]
```

```python
# filtering on the data with multiple values for same column
active_players = df[df['GAMES_STARTED'].isin(['0','1'])]

# filtering on the data with single value for one column
# one way
inactive_players = df[df['GAMES_STARTED']=='Inactive']

# second way
did_not_play = df.query("GAMES_STARTED=='Did Not Play'")


```

```python
active_players['ASSISTS_AND_POINTS'] = active_players['ASSISTS'] + active_players['POINTS']
```

```python
def calculate_time_in_seconds(minutes_played):
    minutes, seconds = minutes_played.split(":")
    total = 60*int(minutes) + int(seconds)
    return total

active_players['minutes_played_in_seconds'] = active_players.apply(
    lambda x: calculate_time_in_seconds(x['MINUTES_PLAYED']), axis=1)

active_players['ASSISTS_AND_POINTS']  = active_players['ASSISTS'] + active_players['POINTS'] + active_players['TOTAL_REBOUNDS']
active_players['EFFICIENCY'] = active_players['ASSISTS_AND_POINTS'] / active_players['minutes_played_in_seconds']
```

```python
active_players =active_players[active_players['minutes_played_in_seconds']>600]
```

```python
active_players.groupby('PLAYER_NAME')[['EFFICIENCY']].max().sort_values('EFFICIENCY', ascending=False)
```
